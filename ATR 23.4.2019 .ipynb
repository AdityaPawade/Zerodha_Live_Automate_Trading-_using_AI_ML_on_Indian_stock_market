# ATR 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data= pd.read_csv("zerodhadata.csv",usecols=["Date","Open","High","Low","Close"])
def EMA(df, column="Close", period=14):
    ema = df[column].ewm(span=period, min_periods=period - 1).mean()
    return df.join(ema.to_frame('EMA'))
a=EMA(data)
pp= ( a["EMA"].shift(1)*13 + data['Close'] ) /  14
print(pp)
pp.plot()
plt.show()
```
