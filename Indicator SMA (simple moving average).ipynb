# Exponential Moving Average (EMA) indicator on SUNPHARMA data ( NSE) 

(from 2019-1-29 to 2019-4-26)

```python
import numpy as np 
import pandas as pd 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#@ opening CSV

df = pd.read_csv("SUNPHARMA1.csv",usecols=["Date","Open","High","Low","Close"])
#@ function 

def SMA(df, column="Close", period=14):
    smavg = df.Close.rolling(14).mean()
    return df.join(smavg.to_frame('SMA'))
a=SMA(df)


#@ plotting 

plt.style.use('seaborn')
a['SMA'].plot()
a['Close'].plot()
plt.show()
print( "   \n \n   \t  \t  SIMPLE MOVING AVERAGE and Close Graph  \n \n ")
print(" \n " )
plt.style.use('seaborn')
a['SMA'].plot()
plt.show()
print( "    \t  \t  SIMPLE MOVING AVERAGE ")
```
