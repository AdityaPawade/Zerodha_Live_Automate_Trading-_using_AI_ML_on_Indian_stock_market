# RSI

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("zerodhadata.csv",usecols=["Date","Open","High","Low","Close"])
def RSI(df, column="Close", period=14):
    delta = df[column].diff()
    up, down = delta.copy(), delta.copy()
    up[up < 0] = 0
    down[down > 0] = 0
    rUp = up.ewm(com=period - 1,  adjust=False).mean()
    rDown = down.ewm(com=period - 1, adjust=False).mean().abs()
    rsi = 100 - 100 / (1 + rUp / rDown)    
    return df.join(rsi.to_frame('RSI'))

a = RSI(data)
print(a)
a['High'].plot()
a['Low'].plot()
a['Close'].plot()
a['Open'].plot()
a["RSI"].plot()
plt.show()
print (" \t' ----- ABOVE GRAPH OF RSI with OHLC data ----- ' \n \n ")
a["RSI"].plot()

plt.show()
print (" \t' ----- ABOVE GRAPH OF only RSI data ----- ' \n \n ")
```
