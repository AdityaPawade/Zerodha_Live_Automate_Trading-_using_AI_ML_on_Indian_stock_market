# OHLC data of IDEA  from (2019-01-23 to 2019-4-22)

```python
df = pd.read_csv("zerodhadata.csv",usecols=["Date","Open","High","Low","Close"])
c=df
print(c)
plt.style.use('seaborn')
df['Close'].plot()
df['High'].plot()
df['Low'].plot()
df['Close'].plot()
plt.show()
print (" \t' ----- ABOVE GRAPH OF OHLC data ----- ' \n \n ")
```

# Close Candles of Idea 

```python
ak=pd.DataFrame(df.Close)
print(ak)
ak['Close'].plot()
plt.show()
print (" \t' ----- ABOVE GRAPH OF is of CLOSE CANDLES----- ' \n \n ")
```

# EMA   [IDEA]

```python
p=ak.Close.ewm(14).mean()
c=pd.DataFrame(p)
print(c)
c['Close'].plot()
plt.show()
print(" \t ' ----- ABOVE GRAPH OF CLOSE EMA-----  '")
```

```python
print ( " \n \n  \n \n \t' ----- COMPARISON between Closed candles and EMA ----- ' .. \n \n " )
ak=pd.DataFrame(df.Close)
ak['Close'].plot()
plt.show()
print (" \t'  ABOVE GRAPH OF is of CLOSE CANDLES ' \n  ")
p=ak.Close.ewm(14).mean()
c=pd.DataFrame(p)
c['Close'].plot()
plt.show()
print(" \t '  ABOVE GRAPH OF CLOSE EMA  '")
```
