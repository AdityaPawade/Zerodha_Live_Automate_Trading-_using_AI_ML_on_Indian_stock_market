# EMA (maths ) 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("zerodhadata.csv",usecols=["Date","Open","High","Low","Close"])
def EMA(df, column="Close", period=14):
    ema = df[column].ewm(span=period, min_periods=period - 1).mean()
    return df.join(ema.to_frame('EMA'))
a=EMA(data)
print (" \t \t  \n \n \t \t  EMA through (maths) \n ")
print(a)
a['Open'].plot()
a['High'].plot()
a['Low'].plot()
a['Close'].plot()
a['EMA'].plot()
plt.show()
a['EMA'].plot()
print (" \t    \t graph only of EMA")
plt.show()

```

# EMA   [IDEA] (pandas)

```python
p=data.Close.ewm(14).mean()
c=pd.DataFrame(p)
print(c)
c['Close'].plot()
plt.show()
print(" \t ' ----- ABOVE GRAPH OF CLOSE EMA through pandas-----  '")
```

```python

df = pd.read_csv("zerodhadata.csv",usecols=["Close"])
print ( " \n \n  \n \n \t' COMPARISON between Closed candles and EMA ' .. \n \n " )
ak=pd.DataFrame(df.Close)
ak['Close'].plot()
plt.show()
print (" \t'  ABOVE GRAPH OF is of CLOSE CANDLES ' \n  ")
p=ak.Close.ewm(14).mean()
c=pd.DataFrame(p)
c['Close'].plot()
plt.show()
print(" \t '  ABOVE GRAPH OF CLOSE EMA  '")
```
